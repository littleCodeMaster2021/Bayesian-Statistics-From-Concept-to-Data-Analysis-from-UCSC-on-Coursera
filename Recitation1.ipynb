## Python Basics

### File format

- Blocks of code are based on indentation

- Spaces and tabs might look similar but are not. Do not use these interchangeably.

### Data Types

 - int, float, list, set, dict, bool, None
 
 - Dynamic typing - Type is determined based on the value assigned and you do have to explicilty define it. It will be changed when a new value of a different type is assigned to the same variable. ** Be careful as these types issues will only be caught during runtime **
 
### Libraries

- Python has lots of libraries like collections, numpy, scipy

- 'import collections' to import the whole library. Use 'collections.Counter()'' in this case to the method

- 'from collections import Counter' to import just the method. Use 'Counter()' in this case

### List and Dictionary Comprehension

- These are used in place of loops and are faster



```python
import time
t0 = time.time()
l = []
for i in range(100000):
  if i%10000 == 0:
    l.append(i)
t1 = time.time()
print(l)
print(t1-t0)
```

```python
t0 = time.time()
l = [i for i in range(100000) if i%10000 == 0]
t1 = time.time()
print(l)
print(t1-t0)
```

## Numpy

```python
#Add an image to the folder where your notebook is located

from matplotlib.pyplot import imread
import matplotlib.pyplot as plt
import numpy as np

butterfly = imread('farfalla1200.jpg')
print(type(butterfly))
print(butterfly.shape)
```

```python
plt.imshow(butterfly)
```

```python
def convert_to_grayscale(image, crop):
    new_img = image[crop:-crop,crop:-crop]
    new_img = np.mean(new_img,axis=2)                
    return new_img
```

```python
butterfly_grayscale = convert_to_grayscale(butterfly, 50)
plt.imshow(butterfly_grayscale, cmap=plt.cm.gray)
```

```python
import math
# sum(set([div*i for div in nums for i in range(1,math.ceil(max_value/div))]))
def sum_multiples(max_value,nums):
        multiples = set()
        for div in nums:
            for i in range(1,math.ceil(max_value/div)):
                multiples.add(div*i)
        return sum(multiples)
```

```python
def sum_fibonacci(min_value,max_value):
    a = 0
    b = 1
    c = a + b
    fibonacci = []
    while c < max_value:
        if c%2 == 0 and c > min_value:
            fibonacci.append(c)
        a = b
        b = c
        c = a + b
    print(fibonacci)
    return sum(fibonacci)
```

```python
def count_subword(subword,words):
    cnt = 0
    for word in words:
        if subword in word and subword != word:
           cnt += 1
    return cnt
```
